## Step 1: Load the "Netflix Movies and TV Shows" Dataset

```python
import pandas as pd
import numpy as np

# Importing "Netflix Movies and TV Shows" dataset
df = pd.read_csv("/content/archive (2).zip")
df.head()
```

## Step 2: Overview

```python
# Check the structure of the dataset
df.info()
```

```python
# Statistical summary
df.describe(include='all')
```

```python
# Check the total number of missing values per column
df.isnull().sum()
```

## Step 3: Handle Missing Values

```python
# Drop rows where 'title' is null
df.dropna(subset=['title'], inplace=True)
```

```python
# Fill missing 'director' with 'Not Available'
df['director'].fillna('Not Available', inplace=True)
```

```python
# Fill missing 'country' with mode
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
# Fill missing 'cast' with 'Not Available'
df['cast'].fillna('Not Available', inplace=True)
```

```python
# Fill missing 'rating' with mode
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
```

```python
df.head()
```

## Step 4: Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

## Step 5: Standardize Text (e.g., country, type)

```python
df['country'] = df['country'].str.strip().str.title()  # Removes extra spaces and standardize country names
df['type'] = df['type'].str.lower()  # Converts all values in the 'type' column to lowercase
```

```python
df.head()
```

## Step 6: Fix Date Formats

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

## Step 7: Rename Columns

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

## Step 8: Check & Fix Data Types

```python
# 'release_year' should be int — check to confirm
df['release_year'] = df['release_year'].astype(int)
```

```python
df.head()
```

## Step 9: Save Cleaned Dataset

```python
df.to_csv("cleaned_netflix_data.csv", index=False)
```
